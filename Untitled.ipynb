## Writer/Director rating dependecy


```python
import re
```

```python
res = []
kek = np.unique(data.loc[:,'writers'])
for i in kek:
    res += re.split(' *, *| *; *| *& *| *story by *', i)
```

```python
writers = {i:[] for i in np.unique(res)[1:]}
```

```python
writers_data = data.loc[:, ['rating', 'votes', 'number_overall']]
```

```python
for i in writers:
    writers_data[i] = 0
```

```python
for n,ws in np.array(data.loc[:, ['number_overall','writers']]):
        print(ws)
```

```python
for n,ws in np.array(data.loc[:, ['number_overall','writers']]):
    res = []
    res += re.split(' *, *| *; *| *& *| *story by *', ws)
    for w in res:
        if w not in writers:
            continue
        writers_data.loc[writers_data['number_overall'] == n, w] = 1
```

```python
X, y = np.array(writers_data.loc[:, writers_data.columns[3:]]),\
        np.array(writers_data.loc[:, 'rating']).astype('float')
#X, y = np.array(word_counts), np.array(data.loc[:, 'rating'])
m = ~np.isnan(y)

X = X[m, :]
y = y[m]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,\
                                                    test_size=0.1)
```

```python
model = lg.LGBMRegressor('rf', learning_rate=0.0001, n_estimators=1000,\
                         colsample_bytree=1, num_leaves=31, subsample=0.8, subsample_freq=1,\
                         reg_alpha=0, min_child_samples=1, min_child_weight=1e-5, importance_type='split')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print(((y_test - y_train.mean())**2).mean())

((y_pred - y_test)**2).mean()
```

```python
explained_variance_score(y_test, y_pred)
```

```python
importances = pd.DataFrame(model.feature_importances_, index=writers_data.columns[3:])\
                            .sort_values(0,ascending=False, axis=0)
```

```python
important_writers = list(importances[:10].index)
```

```python
mask = writers_data.loc[:, important_writers].sum(axis=1) == 1
```

```python
authors = []
ratings = []
for author in important_writers:
    mask = writers_data.loc[:, author] == 1
    authors += [author]*(mask&m).sum()
    ratings += list(writers_data.loc[mask&m,'rating'])
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,10))
sns.stripplot(ratings, authors , 
              size=10,jitter=0.3, edgecolor='none')
sns.despine()
plt.savefig('rating-writer.png', format='png', bbox_inches='tight')
```

## Activity of character to time

```python
import pickle as pk
```

```python
with open('Top_characters', 'rb') as fp:
    top = pk.load(fp)
```

```python
word_counts = pd.DataFrame(0,index = np.arange(data.shape[0]), columns=[character for character in top.index]\
                           + ['season'])
```

```python
for i,(transcript, season) in enumerate(np.array(data.loc[:, ['transcript', 'season']])):
    if type(transcript) is float:
        continue
    for line in eval(transcript):
        if line['character'] in word_counts.columns:
            word_counts.loc[i,line['character']] += len(line['line'].split(' '))
    word_counts.loc[i,'season'] = season 
```

```python
word_counts_ = word_counts.groupby('season').sum(axis=0)
```

```python
word_counts_ = word_counts_.loc[1:, word_counts_.columns[:8]]
word_counts_ = word_counts_.div(word_counts_.sum(1), 0)
```
